## Pendahuluan

Cuaca merupakan salah satu faktor penting yang memengaruhi kehidupan sehari-hari manusia, baik dalam aktivitas sosial, ekonomi, maupun lingkungan. Cuaca adalah suatu fenomena alam yang berkaitan dengan kondisi udara di bumi, setiap harinya cuaca pasti berubah tergantung kondisi udaranya. Secara teknis, cuaca adalah keadaan udara di atmosfer pada waktu dan tempat tertentu yang sifatnya tidak menentu dan berubah-ubah [wikipedia]. Ada banyak faktor yang dapat memengaruhi cuaca, keadaan cuaca dipengaruhi oleh beberapa unsur cuaca yang saling berkaitan satu sama lain yaitu suhu udara, kelembaban udara, kecepatan angin, tekanan udara dan intensitas penyinaran matahari. Menurut Gibbs tahun 1987, cuaca merupakan keadaan atmosfer yang dinyatakan dengan nilai berbagai parameter, antara lain suhu, tekanan, angin, kelembaban dan berbagai fenomena hujan, di suatu tempat atau wilayah selama kurun waktu yang pendek (menit, jam, hari, bulan, musim atau tahun) [laborr].

Prakiraan cuaca yang akurat sangat dibutuhkan untuk membantu masyarakat dalam perencanaan kegiatan, terutama di sektor pertanian, transportasi, dan mitigasi bencana. Dikarenakan banyaknya tuntutan dari berbagai pihak yang membutuhkan informasi kondisi cuaca yang lebih cepat, lengkap, dan akurat maka pemilihan metode yang tepat dalam penentuan cuaca terus dilakukan agar dapat ditemukan metode yang paling tepat dengan tingkat akurasi yang tinggi untuk menentukan prediksi cuaca [adoc.pub]. Prediksi cuaca sering kali sulit dipastikan, keakuratan dalam memprediksinya menjadi salah satu hal utama yang perlu diteliti, terutama dengan adanya perubahan yang sangat fluktuatif. Oleh karena itu, digunakan metode _Fuzzy Logic_ untuk menyelesaikan permasalahan tersebut.

Logika fuzzy adalah logika yang kabur atau
mengandung unsur ketidakpastian [jurnal ilmu komputer]. Logika fuzzy merupakan sistem yang dibangun dengan definisi, cara kerja dan deskripsi yang jelas. Metode logika fuzzy dapat digunakan untuk prediksi cuaca karena logika fuzzy dapat menyelesaikan masalah yang mengandung ketidakpastian, ketidakjelasan, dan ketidaktepatan [FMIPA Univ Lampung Mangkurat].

# Metode dan Eksperimen

## Penjelasan Mengenai Dua Metode yang Digunakan

Dalam penelitian ini, kami menggunakan 2 metode logika fuzzy yaitu metode **Mamdani** dan metode **Sugeno**. Berikut adalah penjelasan dari masing-masing metode:
1. Fuzzy Mamdani

Metode ini diperkenalkan oleh Ebrahim Mamdani pada tahun 1975. Fuzzy model Mamdani sering juga dikenal dengan nama metode max-min. Metode ini menggunakan aturan if-then dengan output yang berbentuk fungsi fuzzy. Tahapan utama dari fuzzy model Mamdani meliputi fuzzifikasi, inferensi fuzzy, komposisi aturan, dan defuzzifikasi.
  - **Fuzzifikasi**: Mengubah input numerik (crisp) menjadi nilai linguistik menggunakan fungsi keanggotaan. Contohnya: Suhu "25°C" dapat diterjemahkan menjadi "Hangat" dengan tingkat keanggotaan tertentu dalam rentang fuzzy.
  - **Inferensi Fuzzy**: Menggunakan aturan logika if-then untuk menentukan output fuzzy berdasarkan input linguistik. Contoh aturan: "If Suhu Hangat AND Kelembapan Tinggi THEN Curah Hujan Ringan."
  - **Komposisi Aturan**: Menggabungkan hasil dari semua aturan fuzzy yang berlaku dengan menggunakan operator logika seperti AND (minimum) atau OR (maximum).
  - **Defuzzifikasi**: Mengubah output fuzzy menjadi nilai numerik (crisp output) menggunakan metode seperti Centroid (pusat massa), Weighted Average, atau Mean of Maximum.
  
  
2. Fuzzy Sugeno

Model Fuzzy Sugeno (model fuzzy TSK) diajukan oleh Takagi, Sugeno, dan Kang (Takagi dan Sugeno, 1985) dalam upaya untuk membangun pendekatan sistematis untuk membangkitkan aturan-aturan fuzzy dari himpunan data input-output yang diberikan. Metode fuzzy Sugeno memiliki penalaran yang hampir sama dengan metode fuzzy Mamdani. Jika output dari fuzzy Mamdani berbentuk fungsi fuzzy, maka output dari fuzzy Sugeno adalah pendekatan yang menghasilkan fungsi matematis atau nilai konstan. Tahapan utama dari fuzzy model Sugeno meliputi fuzzifikasi, inferensi fuzzy, agregasi output, dan defuzzifikasi.
  - **Fuzzifikasi**: Sama seperti Mamdani, input numerik diubah menjadi nilai linguistik menggunakan fungsi keanggotaan. Contoh: Suhu "25°C" diterjemahkan menjadi "Hangat".
  - **Inferensi Fuzzy**: Menggunakan aturan if-then, tetapi outputnya berupa fungsi linier atau nilai tetap. Contoh aturan: "If Suhu Hangat AND Kelembapan Tinggi THEN Curah Hujan = 0.5 × Suhu + 20."
  - **Agregasi Output**: Menggabungkan semua hasil aturan fuzzy menjadi satu nilai crisp output menggunakan fungsi agregasi, misalnya Weighted Average.
  - **Defuzzifikasi**: Data numerik yang telah memiliki derajat keanggotaan diolah kembali menjadi data numerik dengan rumus weight average atau mencari nilai rata-rata dari nilai yang dihasilkan.

### Fuzzifikasi

```python
# Mendefinisikan Input
temperature = ctrl.Antecedent(np.arange(18, 41, 1), 'temp')     # Suhu (°C)
humidity = ctrl.Antecedent(np.arange(70, 101, 1), 'humidity')   # Kelembaban (%)
wind_velocity = ctrl.Antecedent(np.arange(0, 16, 1), 'wind')    # Kecepatan angin (km/jam)

# Mendefiniskan Output
weather = ctrl.Consequent(np.arange(0, 101, 1), 'weather_pred') # Kategori cuaca
```

- Variabel input yang didefinisikan:
  - `temperature`: Suhu udara dengan rentang 18°C sampai 40°C yang diukur dengan kenaikan 1°C.
  - `humidity`: Kelembapan udara mulai dari 70% hingga 100% yang naik per 1%.
  - `wind_velocity`: Kecepatan angin yang berkisar antara 0 sampai 15 km/jam dengan kenaikan 1 km/jam.

- Variabel output didefinisikan sebagai `weather_pred`, yaitu hasil prediksi cuaca. Nilainya ada di rentang 0 sampai 100 yang digunnakan untuk menunjukkan kategori cuaca.

```python
# Fungsi Keanggotaan Temperature
temperature['dingin'] = fuzz.trapmf(temperature.universe, [18, 18, 21, 22])
temperature['normal'] = fuzz.trimf(temperature.universe, [21, 23.5, 26])
temperature['panas'] = fuzz.trapmf(temperature.universe, [25, 26, 40, 40])
```

```python
temperature.view()
```

- `dingin`: Fungsi keanggotaan "dingin" menggunakan bentuk trapezoidal dengan rentang suhu 18°C hingga 22°C.
  - Suhu di bawah 18°C memiliki keanggotaan penuh dalam kategori "dingin".
  - Suhu antara 21°C dan 22°C mengalami transisi menuju nol, yang berarti semakin tinggi suhu, semakin sedikit keanggotaan dalam kategori ini.

- `normal`: Fungsi keanggotaan "normal" menggunakan bentuk triangular dengan rentang suhu 21°C hungga 26°C.
  - Suhu puncaknya ada di sekitar 23.5°C, suhu tersebut adalah yang paling "normal".
  - Suhu di luar 21°C atau 26°C tidak lagi termasuk dalam kategori "normal".

- `panas`: Fungsi keanggotaan "panas" berbentuk trapezoidal dengan rentang suhu 25°C hingga 40°C.
  - Suhu di atas 26°C hingga 40°C sepenuhnya termasuk dalam kategori "panas".
  - Suhu antara 25°C hingga 26°C mulai masuk ke kategori "panas".

```python
# Fungsi keanggotaan Humidity
humidity['kering'] = fuzz.trapmf(humidity.universe, [70, 70, 75, 80])
humidity['lembab'] = fuzz.trimf(humidity.universe, [75, 82.5, 90])
humidity['basah'] = fuzz.trapmf(humidity.universe, [85, 90, 100, 100])
```

```python
humidity.view()
```

- `kering`: Fungsi "kering" menggunakan bentuk trapezoidal dengan rentang kelembapan 70% hingga 80%.

  - Kelembapan antara 75% hingga 80% mulai keluar dari kategori "kering".

- `lembab`: Fungsi keanggotaan "lembab" berbentuk triangular dengan rentang kelembapan 75% hingga 90%.
  - Kelembapan dengan nilai puncak 82.5% memiliki keanggotaan penuh sebagai "lembab".  
  - Semakin jauh dari nilai 75% atau 90%, semakin sedikit keanggotaan dalam kategori "lembab".

- `basah`: Fungsi "basah" menggunakan bentuk trapezoidal dengan rentang kelembapan 85% hingga 100%.
  - Kelembapan di atas 90% hingga 100% sepenuhnya dianggap "basah."  
  - Transisi dari kategori "basah" dimulai di 85%.


```python
# Fungsi keanggotaan Wind Velocity
wind_velocity['pelan'] = fuzz.trapmf(wind_velocity.universe, [0, 0, 3, 3])
wind_velocity['sedang'] = fuzz.trimf(wind_velocity.universe, [3, 4.5, 6])
wind_velocity['kencang'] = fuzz.trapmf(wind_velocity.universe, [6, 6, 15, 15])
```

  - Nilai lebih dari 3 km/jam tidak lagi masuk kategori "pelan".

- `sedang`: Fungsi "sedang" berbentuk triangular dengan rentang kecepatan angin 3 km/jam hingga 6 km/jam.
  - Kecepatan angin dengan  nilai puncak 4.5 km/jam memiliki keanggotaan penuh sebagai "sedang."  
  - Kecepatan di bawah 3 km/jam atau di atas 6 km/jam mulai keluar dari kategori "sedang."

- `kencang`: Fungsi "kencang" menggunakan bentuk trapezoidal dengan rentang kecepatan angin 6 hingga 15 km/jam.
  - Kecepatan angin di atas 6 km/jam hingga 15 km/jam sepenuhnya dianggap "kencang".

```python
# Fungsi keanggotaan Weather
weather['Cerah'] = fuzz.trapmf(weather.universe, [0, 0, 20, 25])
weather['Berawan'] = fuzz.trimf(weather.universe, [20, 35, 50])
weather['Hujan Ringan'] = fuzz.trimf(weather.universe, [40, 57.5, 75])
weather['Hujan Lebat'] = fuzz.trapmf(weather.universe, [70, 80, 100, 100])
```

- `Berawan`: Fungsi ini menggunakan model triangular dengan puncaknya di 35 dari rentang 20 sampai 50.

- `Hujan Ringan`: Fungsi ini menggunakan model triangular dengan puncaknya di 57.5 dari rentang 40 sampai 75.
  - Cuaca dengan nilai sekitar 57.5, sepenuhnya dianggap "Hujan Ringan".
  - Nilai yang lebih kecil dari 40 atau lebih besar dari 75 akan semakin kurang dianggap "Hujan Ringan".

- `Hujan Lebat`: Fungsi keanggotaan untuk kategori ini menggunakan trapezoidal dengan rentang 70 hingga 100.
  - Jika cuaca berada di rentang 70 hingga 100, maka sepenuhnya dianggap "Hujan Lebat".

Tahap inferensi fuzzy pada model Mamdani adalah proses menerapkan aturan-aturan logika fuzzy untuk menentukan output berdasarkan input fuzzy. Setiap aturan menghubungkan kombinasi kategori variabel input (suhu, kelembapan, kecepatan angin) dengan kategori variabel output (cuaca). Proses ini melibatkan evaluasi derajat keanggotaan input menggunakan operator logika fuzzy seperti **AND** (mengambil nilai minimum) untuk menentukan derajat keanggotaan output pada setiap aturan. Hasil dari semua aturan digabungkan menggunakan operator **OR** (mengambil nilai maksimum) untuk menghasilkan output fuzzy akhir, yang kemudian digunakan pada tahap defuzzifikasi.

### Komposisi Aturan

```python
weather_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9,
                                   rule10, rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18,
                                   rule19, rule20, rule21, rule22, rule23, rule24, rule25, rule26, rule27])

weather_sim = ctrl.ControlSystemSimulation(weather_ctrl)
```

- `weather_ctrl`: Mendefinisikan aturan-aturan fuzzy yang telah dibuat dan menghubungkan input dengan output melalui logika fuzzy.
- `weather_sim`: Melakukan simulasi berdasarkan aturan yang ada dan menghasilkan output yang tepat sesuai dengan kondisi input yang diberikan.

### Defuzzifikasi

```python
predicted_weather_list = []
weather_label_list = []

for i, row in df.iterrows():
    try:
        weather_sim.input['temp'] = np.clip(row['Temperature'], 18, 40)
        weather_sim.input['humidity'] = np.clip(row['Humidity(%)'], 70, 100)
        weather_sim.input['wind'] = np.clip(row['Wind_velocity'], 0, 15)
        weather_sim.compute()
        pred_weather = weather_sim.output['weather_pred']
        predicted_weather_list.append(pred_weather)

        if pred_weather <= 20:
            weather_label = 'Cerah'
        elif 20 < pred_weather <= 40:
            weather_label = 'Berawan'
        elif 40 < pred_weather <= 60:
            weather_label = 'Hujan Ringan'
        else:
            weather_label = 'Hujan Lebat'

        weather_label_list.append(weather_label)
    except KeyError:
        predicted_weather_list.append(np.nan)
        weather_label_list.append('Unknown')

df['predicted_weather'] = weather_label_list
```

```python
df['accuracy'] = (df['Actual_Weather'] == df['predicted_weather']).astype(int)
accuracy_percentage = df['accuracy'].mean() * 100
df
```

Nilai fuzzy (prediksi cuaca) yang diperoleh dari sistem diubah menjadi nilai kategori yang jelas dan dapat dimengerti (label cuaca). Output prediksi cuaca `pred_weather` yang merupakan hasil dari sistem fuzzy didefinisikan sebagai berikut:
- Jika `pred_weather` <= 20, cuaca dianggap Cerah.
- Jika 20 < `pred_weather` <= 40, cuaca dianggap Berawan.
- Jika 40 < `pred_weather` <= 60, cuaca dianggap Hujan Ringan.
- Jika `pred_weather` > 60, cuaca dianggap Hujan Lebat.

Akurasi dihitung dengan membandingkan prediksi cuaca yang telah didefinisikan (melalui label) dengan cuaca yang sebenarnya (`Actual_Weather`). Kolom accuracy berisi nilai 1 jika prediksi cuaca benar, dan 0 jika salah.

## Fungsi Keanggotaan

```python
# Fungsi keanggotaan temperature
def temp_dingin(temp_value):
    return np.clip((22 - temp_value) / 4, 0, 1)

def temp_normal(temp_value):
    return np.clip(1 - abs(temp_value - 23.5) / 2.5, 0, 1)

def temp_panas(temp_value):
    return np.clip((temp_value - 25) / 15, 0, 1)

# Fungsi keanggotaan humidity
def humidity_kering(humidity_value):
    return np.clip((80 - humidity_value) / 10, 0, 1)

def humidity_lembab(humidity_value):
    return np.clip(1 - abs(humidity_value - 82.5) / 7.5, 0, 1)

def humidity_basah(humidity_value):
    return np.clip((humidity_value - 85) / 15, 0, 1)

# Fungsi keanggotaan wind velocity
def wind_pelan(wind_value):
    return np.clip((3 - wind_value) / 3, 0, 1)

def wind_sedang(wind_value):
    return np.clip(1 - abs(wind_value - 4.5) / 1.5, 0, 1)

def wind_kencang(wind_value):
    return np.clip((wind_value - 6) / 9, 0, 1)
```




**1. Temperature**

a.) temp_dingin(temp_value): Fungsi ini menghitung derajat keanggotaan suhu terhadap himpunan "dingin". Nilai yang dikembalikan berkisar antara 0 hingga 1, dimana:

* Jika suhu 22°C atau lebih rendah, nilai keanggotaan adalah 1 (suhu sepenuhnya dingin).
* Jika suhu naik, nilai keanggotaan berkurang secara linier, dan menjadi 0 ketika suhu mencapai 26°C.

b.) temp_normal(temp_value):

* Fungsi ini menghitung derajat keanggotaan suhu terhadap himpunan "normal". Fungsi ini mencapai puncaknya pada suhu 23,5°C (nilai keanggotaan 1) dan menurun seiring suhu menjauh dari nilai tersebut. Fungsi ini mengembalikan 1 untuk suhu antara 21°C hingga 26°C dan menurun linier menjadi 0 semakin jauh dari suhu 23,5°C.

c.) temp_panas(temp_value):
*  Fungsi ini menghitung derajat keanggotaan suhu terhadap himpunan "panas". Nilai keanggotaan meningkat linier dari 0 saat suhu 25°C, dan mencapai 1 ketika suhu melebihi 40°C. Semakin panas suhu, semakin tinggi keanggotaan dalam himpunan "panas".

**2.   Humidity**

a.) humidity_kering(humidity_value):
* Fungsi ini menghitung derajat keanggotaan kelembapan terhadap himpunan "kering". Untuk nilai kelembapan 80% atau lebih rendah, nilai keanggotaan adalah 1 (sepenuhnya kering). Seiring kelembapan meningkat, nilai keanggotaan menurun linier, dan menjadi 0 pada kelembapan 90%.

b.) humidity_lembab(humidity_value):
* Fungsi ini menghitung derajat keanggotaan kelembapan terhadap himpunan "lembab". Fungsi ini mencapai puncaknya pada kelembapan 82,5% (nilai keanggotaan 1) dan menurun seiring kelembapan menjauh dari nilai tersebut. Fungsi ini mengembalikan 0 untuk kelembapan yang jauh dari 82,5%.

c.) humidity_basah(humidity_value):
* Fungsi ini menghitung derajat keanggotaan kelembapan terhadap himpunan "basah". Nilai keanggotaan mulai meningkat dari 0 saat kelembapan 85% dan mencapai 1 saat kelembapan melebihi nilai tersebut.a.)

**3. Wind Velocity**

a.) wind_pelan(wind_value):
* Fungsi ini menghitung derajat keanggotaan kecepatan angin terhadap himpunan "pelan". Untuk kecepatan angin antara 0 hingga 3 m/s, nilai keanggotaan meningkat dari 0 menjadi 1. Jika kecepatan angin lebih dari 3 m/s, nilai keanggotaan kembali menjadi 0.

b.) wind_sedang(wind_value):
* Fungsi ini menghitung derajat keanggotaan kecepatan angin terhadap himpunan "sedang". Nilai keanggotaan mencapai puncaknya pada 4,5 m/s (nilai keanggotaan 1) dan menurun linier saat kecepatan angin semakin menjauh dari nilai tersebut.

c.) wind_kencang(wind_value):
* Fungsi ini menghitung derajat keanggotaan kecepatan angin terhadap himpunan "kencang". Nilai keanggotaan meningkat linier dari 0 saat kecepatan angin 6 m/s dan mencapai 1 pada kecepatan angin 15 m/s.



## Inferensi Fuzzy dan Fuzzy Rules

```python
def inferensi(temp_value, humidity_value, wind_value):
    # Derajat keanggotaan
    dingin = temp_dingin(temp_value)
    normal = temp_normal(temp_value)
    panas = temp_panas(temp_value)

    kering = humidity_kering(humidity_value)
    lembab = humidity_lembab(humidity_value)
    basah = humidity_basah(humidity_value)

    pelan = wind_pelan(wind_value)
    sedang = wind_sedang(wind_value)
    kencang = wind_kencang(wind_value)


    # Fuzzy Rules
    rules = [
        dingin * kering * pelan,
        dingin * kering * sedang,
        dingin * kering * kencang,
        dingin * lembab * pelan,
        dingin * lembab * sedang,
        dingin * lembab * kencang,
        dingin * basah * pelan,
        dingin * basah * sedang,
        dingin * basah * kencang,
        normal * kering * pelan,
        normal * kering * sedang,
        normal * kering * kencang,
        normal * lembab * pelan,
        normal * lembab * sedang,
        normal * lembab * kencang,
        normal * basah * pelan,
        normal * basah * sedang,
        normal * basah * kencang,
        panas * kering * pelan,
        panas * kering * sedang,
        panas * kering * kencang,
        panas * lembab * pelan,
        panas * lembab * sedang,
        panas * lembab * kencang,
        panas * basah * pelan,
        panas * basah * sedang,
        panas * basah * kencang,
    ]

     # Fungsi linier
    def linear_output(temp_value, humidity_value, wind_value):
        return 0.3 * temp_value + 0.4 * humidity_value + 0.3 * wind_value
    output = sum(rule * linear_output(temp_value, humidity_value, wind_value) for rule in rules) / max(sum(rules), 1e-6)

    return output
```

## Defuzzifikasi

```python
results = []
for _, row in df.iterrows():
    weather_quality = inferensi(row['Temperature'], row['Humidity(%)'], row['Wind_velocity'])
    results.append({
        'Day' : row['Day'],
        'Temperature': row['Temperature'],
        'Humidity': row['Humidity(%)'],
        'Wind_velocity': row['Wind_velocity'],
        'Weather Quality': weather_quality
    })


output_sugeno = pd.DataFrame(results)
output_sugeno
```

Weather Quality memiliki rentang nilai yang dipengaruhi oleh input variabel. Semakin tinggi nilai Weather Quality, maka cuaca bisa dikategorikan pertanda hujan. sebagai contoh:

* Cuaca dengan suhu panas (misalnya 31°C) dan kelembapan rendah (64%) menghasilkan Weather Quality yang sangat rendah (0.00), yang dapat menandakan cuaca yang baik atau cerah.

* Cuaca dengan suhu normal (24-28°C) dan kelembapan tinggi (95%) cenderung menghasilkan Weather Quality yang lebih tinggi (sekitar 45 hingga 47), yang menunjukkan kondisi cuaca yang buruk atau bisa dikatakan hujan.
